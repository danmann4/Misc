# Worker Widget Data Generation

This is a generic take home question to generate some data and process it from Towards Data Science. 
The first script "Workers_Widgets" generates said data, while this script processes the data using various python packages.

The idea for the data is to generate a workers dataframe with various employee details (full/part time, Team, Status, etc.). Accompanying this dataframe is another that has product ("widget") information like an item number, the worker that made it, and the time taken at various steps to manufacture the widget.

We will disect the attributes of the two data frames and try to demonstrate a typical approach to cleaning/visualizing an unknown dataset. 

```python
#import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#import the csvs
worker_df = pd.read_csv('data_workers.csv')
widget_df = pd.read_csv('data_widgets.csv')
```

```python
#print both dataframes to see what the rows and cols are
print(worker_df.head())
print(worker_df.dtypes)
```

```python
print(widget_df.head())
print(widget_df.dtypes)
```

```python
#Lets check for the number of unique values in each column, and see if there is any NAs in the data. 
print('Worker_df')
for col in worker_df.columns:
    print(f'Unique {col}: {worker_df[col].nunique()}')
print(f'\nNAs:\n{pd.isna(worker_df).any()}')
print(f'\nFile shape: {worker_df.shape}\n')

print('Widget_df')
for col in widget_df.columns:
    print(f'Unique {col}: {widget_df[col].nunique()}')
print(f'\nNAs:\n{pd.isna(widget_df).any()}')
print(f'\nFile shape: {widget_df.shape}')
```

```python
#We know there isn't much cleaning to do since we generated the data. However, if we wanted to balance data sets or remove NAs we would do it here. 
#Lets now check out some summary stats for the columns to get a feel for the data.
print('Worker_df')
print(worker_df.describe())

print('\nWidget_df')
print(widget_df.describe())
```

## Data at a glance

So summary stats of continuous variables in the worker df predictably aren't very useful (since its just worker ID), but the widget df is. We can see a rough ballpark of the average time, and spread in time required for each widget step. However, mean and SD aren't super useful yet since we haven't even proven if the steps are normally distributed. We can tell from the mean and 75% quantile though that the step 3 appears to have a larger proportion values skewed towards longer times than say step 1.

Next up, we just want to convert our df columns into the proper data type, like datetime for 'Hire Date'.

```python
#Convert dates
worker_df['Hire Date'] = pd.to_datetime(worker_df['Hire Date'], format='%Y-%m-%d')

worker_df.info()
worker_df.head()
```

## Data Visualization

Now we can start visualizing the data we are working with and get a better grasp of what we have on our hands. 
First off we can start to dissect the simpler worker df. We can check for things like the most common hire dates, most popular teams, and get a better feel for where the workers are distributed in this dataset. 

```python
# Let check out the most common hire dates. We can make a counter for each one and make a plot of the top 15. 
hire_dates = Counter(worker_df['Hire Date'].dt.date)
common_dates = [d[0] for d in hire_dates.most_common(15)]
common_counts = [d[1] for d in hire_dates.most_common(15)]
# This is just a helper function to make our plot a little more readable, its going to place the count for each
# date on top of the bar in the bar graph. 

# function to show values on bars
def show_values_on_bars(axs):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()
            value = '{:.0f}'.format(p.get_height())
            ax.text(_x, _y, value, ha="center")
    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)

# plot most common hire dates
fig, ax = plt.subplots()
g = sns.barplot(common_dates, common_counts, palette='colorblind')
g.set_yticklabels([])
ax.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False) # labels along the bottom edge are off
#Show the values on top of the bars:
show_values_on_bars(ax)

sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('')
plt.title('Most Common Hire Dates', fontsize=30)
plt.tick_params(axis='x', which='major', labelsize=12)
fig.autofmt_xdate()
plt.show()
```

```python
#Lets do the same thing again, but for the most common years of hire
hire_years = Counter(worker_df['Hire Date'].dt.year)
common_years = [d[0] for d in hire_years.most_common()]
common_counts = [d[1] for d in hire_years.most_common()]

fig, ax = plt.subplots(figsize = (10, 5))
g = sns.barplot(common_years, common_counts, palette = 'colorblind')
g.set_yticklabels([])
ax.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False) # labels along the bottom edge are off

show_values_on_bars(ax)

sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('')
plt.title('Workers by Year Hired', fontsize=30)
ax.tick_params(axis='x', which='major', labelsize=10)
fig.autofmt_xdate()
plt.show()
```

```python
# Nothing extremely informative from these plots so far, we see that the most common dates are quite randomly distributed with
# no clear trend. Lets now check out the worker status using seaborn for the plotting. 

fig, ax = plt.subplots(figsize = (5, 5))
g = sns.countplot(x="Worker Status", data = worker_df, palette='colorblind')
g.set_yticklabels([])
ax.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False) # labels along the bottom edge are off

# Lets take advantage agaain of that function we defined to show the value on top of the bars. 
show_values_on_bars(ax)
plt.title('Workers by Worker Status')
sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('')
plt.tick_params(axis='x', which='major', labelsize=12)
plt.show()
```

```python
# So the last plot clearly shows the bulk of workers at this company are full time. 
# Could be useful info depending on what our needs are. 

# visualize team features now.
sns.countplot(y='Team', data = worker_df, palette='colorblind')
sns.despine(left=True, bottom=True)
plt.title('Workers by Team')
plt.show()
```

```python
# Lets merge our last two plots and check on how the teams are distributed by worker status. 
fig, ax = plt.subplots(figsize = (6,3))
g = sns.countplot(x='Team', data = worker_df, hue='Worker Status', palette='colorblind')
g.set_yticklabels([])
ax.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False) # labels along the bottom edge are off

show_values_on_bars(ax)

# position the legend so that it doesn't cover any box
ax.legend(bbox_to_anchor=(1.05, 1.05),
          ncol=1, fancybox=True, shadow=True)
 

sns.despine(left=True, bottom=True)
plt.title('Workers by Team and Worker Status')
plt.ylabel('')
plt.xlabel('')
plt.show()
```

## Worker Dataframe

From our worker df we were able to glean simple things, like a majority of workers being full time. 
We were also able to learn that there doesn't seem to be any trends in hire dates, or the distribution of workers across teams. 

Now lets look at the widget df. 

```python
# As we hinted at earlier, lets check out the distribution of times required for each widget step. 
# This will tell us what statistics we should be more concerned with . 

g = sns.kdeplot(data=widget_df[['Step 1', 'Step 2', 'Step 3']], fill = True, 
           alpha = 0.5)
plt.xlim([0, 30])
```

## Widget Dataframe

We can see from the plot above each step follows a very different distribution. We can see now why we picked up a larger portion of higher times for step 3 compared to 1, since it looks to be exponentially distributed (playing dumb, we know its exponential). We see step 2 looks normally distributed, while step 1 could either be log-normal or some gamma distribution. 

From here we could do something like an Anderson-Darling or Shapiro-Wilk test to confirm our distribution assumptions, but we know exactly what they are since we made them! So lets move onto joining the data frames to create a more complete table of all the results. 

```python
merged_df = pd.merge(worker_df, widget_df, how = 'inner', on = 'Worker ID')
print(merged_df.shape)
merged_df.head()
```

```python
# Now that we merged the data frames we can do things like linking names to widget counts. 
# Lets make a plot of the workers that have the lowest amount of widgets produced. We'll
# simply just rank them by the lowest 20 and make a bar graph of it.

# first create temporary df 
tmp = merged_df.groupby(['Worker Name']).count().sort_values(by='Worker ID', ascending=True)['Item Number']
fig, ax = plt.subplots(figsize = (10,5))
vals = tmp.values.flatten()[:20]

g = sns.barplot(x = vals, y = tmp.index[:20],
                palette = 'colorblind', ax=ax)
                       
plt.title('Workers with Lowest Item Count')
sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('')
plt.tick_params(axis='x', which='major', labelsize=12)
plt.show()
```

```python
# One of the more fun plots to make is a violin plot for visual comparison of two distributions. 
# Lets look at distribution of time required for each step by worker status (part/full time). 

tmp = merged_df.loc[merged_df['Worker Status'].isin(
    ['Full Time','Part Time'])] # grab each row associated with the status we are interested in
# list of steps to loop over
steps = ['Step 1', 'Step 2', 'Step 3']
# create a plot for each step
for step in steps:
    fig, ax = plt.subplots(figsize=(10, 5))
    g = sns.violinplot(x='Team', 
                   y=step, 
                   hue='Worker Status', 
                   split=True, data=tmp)
    sns.despine(left=True, bottom=True)
    plt.xlabel('')
    plt.ylabel('Time')
    plt.title(f'{step} by Worker Status', fontsize=20)
    plt.show()
```
